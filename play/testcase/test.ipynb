```python
import micom
from micom import Community
from cobra import Reaction, Metabolite

# Check MICOM version
print("MICOM version:", micom.__version__)

# Load individual models
model_bin5 = micom.util.load_model('bin.5.xml')
model_bin6 = micom.util.load_model('bin.6.xml')

# Define the MF2 medium
medium = {
    'EX_cpd00058_e0': 1000.0,   # Cu2+
    'EX_cpd00104_e0': 1000.0,   # Biotin
    'EX_cpd00063_e0': 1000.0,   # Ca2+
    'EX_cpd00099_e0': 1000.0,   # Cl-
    'EX_cpd00149_e0': 1000.0,   # Cobalt
    'EX_cpd10515_e0': 1000.0,   # Fe2+
    'EX_cpd10516_e0': 1000.0,   # Fe3+
    'EX_cpd00067_e0': 1000.0,   # H+
    'EX_cpd00001_e0': 1000.0,   # H2O
    'EX_cpd00242_e0': 1000.0,   # H2CO3
    'EX_cpd00205_e0': 1000.0,   # K+
    'EX_cpd00254_e0': 1000.0,   # Mg
    'EX_cpd00030_e0': 1000.0,   # Mn2+
    'EX_cpd11574_e0': 1000.0,   # Molybdate
    'EX_cpd00244_e0': 1000.0,   # Nickel
    'EX_cpd00209_e0': 1000.0,   # Nitrate
    'EX_cpd00009_e0': 1000.0,   # Phosphate
    'EX_cpd00048_e0': 1000.0,   # Sulfate
    'EX_cpd00305_e0': 1000.0,   # Thiamin
    'EX_cpd00034_e0': 1000.0,   # Zn2+
    'EX_cpd11632_e0': 1000.0,   # Photon
    'EX_cpd00011_e0': 1000.0,   # CO2
    'EX_cpd00007_e0': 1000.0,   # O2
    'EX_cpd00528_e0': 1000.0,   # N2
}

# Function to adjust medium for a model
def adjust_medium(model, medium):
    adjusted_medium = {}
    for rxn_id, value in medium.items():
        if rxn_id in [rxn.id for rxn in model.exchanges]:
            adjusted_medium[rxn_id] = value
        else:
            print(f"Exchange reaction {rxn_id} not found in model {model.id}")
    return adjusted_medium

# Function to add a metabolite to a model
def add_metabolite_to_model(model, metabolite_id, formula, name, compartment):
    metabolite = Metabolite(
        id=metabolite_id,
        formula=formula,
        name=name,
        compartment=compartment
    )
    model.add_metabolites([metabolite])
    print(f"Added metabolite {metabolite_id} to model {model.id}")

# Function to add an exchange reaction to a model
def add_exchange_reaction(model, metabolite_id, reaction_id=None, lower_bound=-1000, upper_bound=1000):
    metabolite = model.metabolites.get_by_id(metabolite_id)
    exchange_rxn = Reaction(id=reaction_id or f"EX_{metabolite.id}")
    exchange_rxn.name = f"Exchange for {metabolite.name}"
    exchange_rxn.lower_bound = lower_bound  # Allow uptake
    exchange_rxn.upper_bound = upper_bound  # Allow secretion
    exchange_rxn.add_metabolites({metabolite: -1})
    model.add_reactions([exchange_rxn])
    print(f"Added exchange reaction {exchange_rxn.id} to model {model.id}")

# Add missing extracellular metabolites to model_bin5
add_metabolite_to_model(
    model_bin5,
    metabolite_id='cpd00104_e0',
    formula='C10H16N2O3S',  # Biotin formula
    name='Biotin',
    compartment='e0'
)
add_metabolite_to_model(
    model_bin5,
    metabolite_id='cpd00242_e0',
    formula='CH2O3',  # Carbonic acid formula
    name='Carbonic acid',
    compartment='e0'
)
# Add missing extracellular metabolites to model_bin6
add_metabolite_to_model(
    model_bin6,
    metabolite_id='cpd00104_e0',
    formula='C10H16N2O3S',  # Biotin formula
    name='Biotin',
    compartment='e0'
)
add_metabolite_to_model(
    model_bin6,
    metabolite_id='cpd00242_e0',
    formula='CH2O3',  # Carbonic acid formula
    name='Carbonic acid',
    compartment='e0'
)

# Add exchange reactions to model_bin5
add_exchange_reaction(
    model_bin5,
    metabolite_id='cpd00104_e0',
    reaction_id='EX_cpd00104_e0'
)
add_exchange_reaction(
    model_bin5,
    metabolite_id='cpd00242_e0',
    reaction_id='EX_cpd00242_e0'
)

# Add exchange reactions to model_bin6
add_exchange_reaction(
    model_bin6,
    metabolite_id='cpd00104_e0',
    reaction_id='EX_cpd00104_e0'
)
add_exchange_reaction(
    model_bin6,
    metabolite_id='cpd00242_e0',
    reaction_id='EX_cpd00242_e0'
)

# Test individual models
for model, name in zip([model_bin5, model_bin6], ['bin.5', 'bin.6']):
    # Adjust medium for the model
    adjusted_medium = adjust_medium(model, medium)
    # Set the medium
    model.medium = adjusted_medium
    # Optimize the model
    sol = model.optimize()
    print(f"Model {name} growth rate: {sol.objective_value}")
    # If growth rate is zero, consider adding missing exchanges or checking the model


```
